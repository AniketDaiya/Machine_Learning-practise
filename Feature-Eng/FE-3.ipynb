```python
# FE
# excersice-3 ( clustering using K-means )

# k-means is a clustering algorithm belonging to unsupervised machine learning algorithms

# it determines the centroids and then divides the points of data ( based on how close distant form the centroid are they )

# n_clusters , defines the no of clusters 
# max_iter , defines the maximum no of iterations ( how k-means works : 1.defines the centroid & adds data points to the nearest centroid's cluster
#  -> 2. redefines the centroid {to minimize the distance to points})

# n_init , deifines the number of iteration of the cycle 

import pandas as pd

data = pd.read_csv('file_path')

# how to use it to create new features ? ( it's sensitive to scale of data ; so often requires normalization / Standardization )

from sklearn.cluster import KMeans

X = data.copy()
y = X.drop("prediction_column")

features = ["MedInc", "Latitude", "Longitude"]

# example of standardization ( here we are trying to make clusters of MedInc acc to lon and lat ; but we won't do this there cuz it will alter the long and lat vals ) 

X_scaled = X.loc[:,features]
X_scaled = (X_scaled - X_scaled.mean(axis = 0 ))/X_scaled.std(axis = 0)

kmeans = KMeans(n_clusters = 5 , random_state= 0)

# we use fit_predict ( to get the clusters { around centroids } in the new column)

X['Clusters'] = kmeans.fit_predict(X)

# we use fit_transform ( to get the distance of points/features form the centroid { of it's cluster } )

X["Clusters_points"] = kmeans.fit_transform(X)




```
