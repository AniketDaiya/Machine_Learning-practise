```python
# FE
# excercise- 1 (mutual information)

# mutual information , is a utility matrix to determine features with high corelation ( not limited to linear ) with the target ; should use data visualization here.

# mutual information treats descrete ( categorical ) and continous ( float , int ) features diffrently , so we need to convert descrete - > continous ( label encoding )

# after finding the high MI features you can also by domain knowledge try to determine interaction effect (between 2 features) ( that can be furtur used to define synthetic features)

# importing data 

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

data = read_csv('file_path')
X = data.copy()
y = X.drop('Price')

# label encoding for categoricals 
for col in X.select_dtypes('object'):
    x[col],_ = x[col].factorize()

# all discrete should now have int data type ( double check before MI)
discrete_features = X.dtypes == int

# now calculating the mutual information 

from sklearn.feature_selection import mutual_info_regression

def make_mi_scores (X,y,discrete_features):
    mi_scores = mutual_info_regression( X,y,discrete_features=discrete_features,random_state = 0 )
    mi_scores = pd.Series(mi_scores , name = 'mi_scores' , index = X.columns)
    mi_scores = mi_scores.sort_values(ascending = False)

    return mi_scores

mi_scores = make_mi_scores(X,y,discrete_features)
mi_scores[::3] # start : stop : step

# visualization of mi_scores 
def plot_mi_scores(scores):
    scores = scores.sort_values(ascending = True)
    width = np.arange(len(scores))
    ticks = list(scores.index)
    plt.barh(width , scores)
    plt.yticks(width , ticks)
    plt.title('Mutual Information Scores')

plt.figure(dpi =100 , figsize = (8,5) )# dots per inch = dpi = resolution 
plot_mi_scores(mi_scores)

                      
```
