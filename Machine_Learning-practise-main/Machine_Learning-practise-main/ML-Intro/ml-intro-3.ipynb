```python
# Machine_Learning-practice
# excercise-3 (model validation)

import pandas as pd
filepath="/kaggle/input/datasets/dansbecker/melbourne-housing-snapshot/melb_data.csv"
home_data = pd.read_csv(filepath)

y=home_data.Price

feature_list=["Rooms","Distance","Postcode"]
X=home_data[feature_list]

# train_test_split ; used to save model from insample score , by spliting the same sample data into training data (to fit the model) and validation data (to evaluate the model)
# this is done before defining and fitting the model ofc!

from sklearn.model_selection import train_test_split
#first break the data 
train_X,val_X,train_y,val_y = train_test_split(X,y,random_state=1)

# so therefore defining and fitting/training the model
from sklearn.tree import DecisionTreeRegressor
home_model = DecisionTreeRegressor(random_state=2)

home_model.fit(train_X,train_y)

predictions = home_model.predict(val_X)

# mean_absolute_error ; function is used to determine the average deviation of predictied values from actual values 
from sklearn.metrics import mean_absolute_error
mean_absolute_error(val_y,predictions)


```
