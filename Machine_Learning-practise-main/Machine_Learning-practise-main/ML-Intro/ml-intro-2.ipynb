```python
# Machine_Learning-practice
# excersice-2 (first machine learning model)

# read the data
import pandas as pd
filepath="/kaggle/input/datasets/dansbecker/melbourne-housing-snapshot/melb_data.csv"
model_data = pd.read_csv(filepath)

# select prediction variable and features ( from intution for now !)

# prediction variable (y) , selected using "."syntax
y = model_data.Price

# features (X) , selected using the list syntax
feature_list=["Rooms","Distance","Postcode"]
X = model_data[feature_list]

# using scikit learn to build a ml model ( it's a very popular lib for modeling data in DataFrames )
from sklearn.tree import DecisionTreeRegressor

# define the type of model and set randam_state , fr reproducibility ( so that while training same output for same input !)
model = DecisionTreeRegressor(random_state=1)

# fit ; recognizing patterns form previous data
model.fit(X,y)

# predict ; finally make predictions 
prediction = model.predict(X)
print(prediction)
```
