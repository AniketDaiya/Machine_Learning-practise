```python
# ML-Inter
# excercise-5 ( cross-validation , better measure of model quality ( with added benifit of cleaning our code ; as no need to keep track of test and train data seperately))

# we use cross validation for small datasets ; because we are only measuring model quality on limited data ( the validation set is limit to specific rows ) 
# because it takes longer time { it estimates multiple models , one for each fold }

import pandas as pd

data = pd.read_csv('file_path')
y = data.Prices
X = data.drop(['Prices'] , axis = 1 )


# use of pipeline make the process a lot simpler

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestRegressor

my_pipeline = Pipeline(steps=[
    ('preprocessor' , SimpleImputer()),
    ('model' , RandomForestRegressor(n_estimatorss = 100 , random_state = 0 ) )
])

# we use the bulit-in function cross_val_score from scikitlearn ( it returns negerive mean abs error ; so making it positive )

from sklearn.model_selection import cross_val_score 

# here cv defines the number of folds the data is divided into for cross-validation ( where each fold is validated once)

score = -1 * cross_val_score(my_pipeline , X , y , cv = 5 , scoring ='neg_mean_absolute_error') # list of MEA for each fold

print('scores for all 5 foldes : ' , score )


# cross-validation can also be used for selecting appropriate parameters in ml models ( by compairing cross validated score for different parameters)
# here we will see the n_estimarots parameter

def get_score(n_estimators):
  
    my_pip = Pipeline(steps=[
      ('preprocessing' , SimpleImputer()),
      ('model' , RandomForestRegressor(n_estimators= n_estimators , random_state=0))
    ])
   
    score = -1 * cross_val_score(my_pip , X , y , cv = 5 , scoring ='neg_mean_absolute_error')
    
    return score.mean()


for i in range(1,9):
    
    print('for {} number of trees , the MEA is : {} '.format(50*i,get_score(50*i))) # for trees 50 , 100 , 150 ..., 450


```
