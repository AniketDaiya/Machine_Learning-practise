```python
# ML-Inter
# excercise-6 gradient boosting (XG Boost , uses ensemble of models with the loss function working on gradient decent 
#  highly accurate for tabular data {if tuned with parameters}

import pandas as pd
from sklearn.model_selection import train_test_split
data = pd.read_csv('file_path')
y = data.Prices
X = data.drop(['Prices'] , axis = 1)
X_train , X_val , y_train , y_val = train_test_split(X,y,train_size=0.8,test_size=0.2,random_state=0)

# one_hot_encoding using pandas 

X_train = pd.get_dummies(X_train) # get_dummies does the entire onehot endcoding process 
X_val = pd.get_dummies(X_val)
X_train , X_val = X_train.align(X_val , join='left' , axis = 1 ) # this makes the ordering and naming of columns exactly same in both dataframes 
# join with left:X_train ( make sure to change the names and add thing that are not in right:X_val)


# using the XG Boost library for an XGBRegressor model

from xgboost import XGBRegressor

model = XGBRegressor( random_state = 0 )
model.fit(X_train , y_train)
predict = model.predict(X_val , y_val)


# now using parameters with XGBRegressor model to improve accuracy of the model 

# n_estimators ; defines the number of models in the ensemble ( simi to trees in randomforest )

# learning_rate ; multiplies each models predictions with a number 
# ( preferably small , so that each tree we add to the ensemble helps us less. So, we can set a higher value for n_estimators without overfitting) by default 0.1

# early_stopping_rounds ; used to determine the optimal number of trees ( with rounds , each round reducing the error )
# eval_set ; used with early_stopping_rounds ( to specify the evalutions set)

# n_jobs ; used for parallelism to build the model faster

model_1 = XBGRegressor(n_estimators = 100 , learning_rate = 0.05)

model_1.fit(X_train , y_train ,
            early_stopping_rounds = 5,
            eval_set = [(X_val , y_val)] ,
            verbose = False) # verbose = 0 ; no output/logs is shown during training


```
